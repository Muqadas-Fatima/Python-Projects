```python
import requests
import tkinter as tk
from tkinter import ttk, messagebox

# Your API Key (Replace with your key)
API_KEY = "e4a042700174fc9216f1f62f"
BASE_URL = f"https://v6.exchangerate-api.com/v6/{API_KEY}/latest/"

def convert_currency():
    try:
        amount_text = amount_entry.get().strip()
        
        if not amount_text:
            messagebox.showerror("Error", "Please enter an amount.")
            return
        
        # Replace commas with dots (if user enters 1,500 instead of 1.500)
        amount_text = amount_text.replace(",", ".")
        
        if not amount_text.replace(".", "").isdigit():
            messagebox.showerror("Error", "Invalid amount. Please enter a valid number.")
            return

        amount = float(amount_text)
        from_currency = from_currency_combo.get()
        to_currency = to_currency_combo.get()

        if not from_currency or not to_currency:
            messagebox.showerror("Error", "Please select both currencies.")
            return
        
        # Fetch exchange rates
        response = requests.get(BASE_URL + from_currency)
        data = response.json()

        if response.status_code != 200:
            messagebox.showerror("Error", "Failed to fetch exchange rates. Try again later.")
            return

        exchange_rates = data.get("conversion_rates", {})
        rate = exchange_rates.get(to_currency)

        if rate is None:
            messagebox.showerror("Error", "Exchange rate not available for this currency.")
            return
        
        converted_amount = amount * rate
        result_label.config(text=f"{amount} {from_currency} = {converted_amount:.2f} {to_currency}")

    except ValueError:
        messagebox.showerror("Error", "Invalid amount. Please enter a valid number.")
    except Exception as e:
        messagebox.showerror("Error", f"Conversion failed: {str(e)}")

# Create main window
root = tk.Tk()
root.title("Currency Converter")
root.geometry("450x350")
root.resizable(False, False)

# Title label
title_label = tk.Label(root, text="Currency Converter", font=("Arial", 16, "bold"))
title_label.pack(pady=10)

# Amount entry
amount_label = tk.Label(root, text="Enter Amount:")
amount_label.pack()
amount_entry = tk.Entry(root)
amount_entry.pack(pady=5)

# List of currencies
currencies = [
    "USD",  # United States Dollar 🇺🇸
    "PKR",# Pakistani 
    "EUR",  # Euro (Used in many European countries) 🇪🇺
    "GBP",  # British Pound Sterling 🇬🇧
    "INR",  # Indian Rupee 🇮🇳
    "JPY",  # Japanese Yen 🇯🇵
    "CAD",  # Canadian Dollar 🇨🇦
    "AUD",  # Australian Dollar 🇦🇺
    "CHF",  # Swiss Franc 🇨🇭
    "CNY",  # Chinese Yuan 🇨🇳
    "SGD",  # Singapore Dollar 🇸🇬
    "PKR",  # Pakistani Rupee 🇵🇰
    "SAR",  # Saudi Riyal 🇸🇦
    "AED",  # UAE Dirham 🇦🇪
    "ZAR",  # South African Rand 🇿🇦
    "MYR",  # Malaysian Ringgit 🇲🇾
    "RUB",  # Russian Ruble 🇷🇺
    "THB",  # Thai Baht 🇹🇭
    "KRW",  # South Korean Won 🇰🇷
    "HKD",  # Hong Kong Dollar 🇭🇰
    "IDR",  # Indonesian Rupiah 🇮🇩
    "BRL",  # Brazilian Real 🇧🇷
    "NGN",  # Nigerian Naira 🇳🇬
    "PHP",  # Philippine Peso 🇵🇭
    "EGP",  # Egyptian Pound 🇪🇬
    "BDT",  # Bangladeshi Taka 🇧🇩
    "MXN"   # Mexican Peso 🇲🇽
]


# Currency dropdowns
from_currency_label = tk.Label(root, text="From Currency:")
from_currency_label.pack()
from_currency_combo = ttk.Combobox(root, values=currencies, state="readonly")
from_currency_combo.pack(pady=5)
from_currency_combo.current(0)  # Default to USD

to_currency_label = tk.Label(root, text="To Currency:")
to_currency_label.pack()
to_currency_combo = ttk.Combobox(root, values=currencies, state="readonly")
to_currency_combo.pack(pady=5)
to_currency_combo.current(1)  # Default to EUR

# Convert button
convert_button = tk.Button(root, text="Convert", command=convert_currency, bg="blue", fg="white")
convert_button.pack(pady=10)

# Result label
result_label = tk.Label(root, text="", font=("Arial", 12, "bold"))
result_label.pack(pady=10)

# Run the application
root.mainloop()
```
