```python
import tkinter as tk
from tkinter import ttk, messagebox
from sympy.physics.units import (
    meter, kilometer, centimeter, millimeter, mile, inch, foot, yard,
    gram, kilogram, milligram, pound,
    second, minute, hour, day, year, convert_to
)

# Dictionary for unit mapping
unit_dict = {
    "Meter (m)": meter, "Kilometer (km)": kilometer, "Centimeter (cm)": centimeter, "Millimeter (mm)": millimeter,
    "Mile (mi)": mile, "Inch (in)": inch, "Foot (ft)": foot, "Yard (yd)": yard,
    "Gram (g)": gram, "Kilogram (kg)": kilogram, "Milligram (mg)": milligram, "Pound (lb)": pound,
    "Second (s)": second, "Minute (min)": minute, "Hour (h)": hour, "Day": day, "Year": year
}

# Conversion function
def convert_units():
    try:
        value = float(value_entry.get())  # Get user input value
        from_unit = unit_dict[from_unit_var.get()]
        to_unit = unit_dict[to_unit_var.get()]
        
        # Convert using SymPy
        converted_value = convert_to(value * from_unit, to_unit)

        # Display result
        result_label.config(text=f"Converted Value: {converted_value}")
    
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number.")
    except KeyError:
        messagebox.showerror("Error", "Invalid unit selection.")

# GUI Setup
root = tk.Tk()
root.title("Unit Converter")
root.geometry("400x300")
root.resizable(False, False)

# Styling
style = ttk.Style()
style.configure("TButton", font=("Arial", 12))
style.configure("TLabel", font=("Arial", 12))
style.configure("TEntry", font=("Arial", 12))

# Heading
ttk.Label(root, text="📏 Unit Converter", font=("Arial", 14, "bold")).pack(pady=10)

# Value Input
ttk.Label(root, text="Enter Value:").pack()
value_entry = ttk.Entry(root, justify="center")
value_entry.pack(pady=5)

# From Unit Selection
ttk.Label(root, text="From Unit:").pack()
from_unit_var = tk.StringVar(value="Meter (m)")
from_unit_dropdown = ttk.Combobox(root, textvariable=from_unit_var, values=list(unit_dict.keys()), state="readonly")
from_unit_dropdown.pack(pady=5)

# To Unit Selection
ttk.Label(root, text="To Unit:").pack()
to_unit_var = tk.StringVar(value="Kilometer (km)")
to_unit_dropdown = ttk.Combobox(root, textvariable=to_unit_var, values=list(unit_dict.keys()), state="readonly")
to_unit_dropdown.pack(pady=5)

# Convert Button
convert_btn = ttk.Button(root, text="Convert", command=convert_units)
convert_btn.pack(pady=10)

# Result Label
result_label = ttk.Label(root, text="Converted Value: ", font=("Arial", 12, "bold"))
result_label.pack(pady=5)

# Run GUI
root.mainloop()
```
